```python


start = time.time()
duration = 0

clock = input('how many seconds you want to do this?  ....')

while duration < int(clock):
    duration = time.time() - start
    a = s.position()
    time.sleep(2)
    b = s.position()
    
    # if after 2 seconds position has changed, print the position.
    if b != a:
       duration = time.time() - start
       selection = s.confirm(text='First try?', title='', buttons=['OK', 'Cancel'])
       
       # record coordinates if confirmed
       if selection == 'OK':
            print ('Recorded')
       else:
        # try recording the next position
            a = s.position()
            time.sleep(2)
            b = s.position()
            if b != a :
                selection = s.confirm(text='Second try?', title='', buttons=['OK', 'Cancel'])
    
        
print ('Session finishes, run again')
```

```python
import pyautogui as s
import time

# ask users how many coordinates to record
# record those coordinates to dataframe

for i in range(2):

    selection = s.confirm(text='Select coordinate', title='', buttons=['OK', 'Cancel'])

    if selection == 'OK':
        time.sleep(4)  # give user time to make selection
        s.alert(text='Saved Succesfully!' + str(i) , title='' , button='OK')
        typ = s.confirm(text='Is this a click?',buttons=['Yes','No'])
        print (s.position(), typ)
    else:
        print ('No record')
```
